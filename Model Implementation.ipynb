### STEP 1 — Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

### STEP 2 — Load Processed Dataset

```python
df = pd.read_csv("stock_datasets/RELIANCE_processed.csv")

print(df.head())
```

### STEP 3 — Check Dataset Info

```python
print(df.info())

print(df.shape)
```

### STEP 4 — Separate Features and Target
#### Features (X)

Everything except Target.

#### Target (y)

Next-day closing price.

```python
X = df.drop(columns=["Target"])

y = df["Target"]
```

### STEP 6 — Train-Test Split
For stock prediction:

```python
split_index = int(len(df) * 0.8)

X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]
```

### STEP 7 — Verify Split Shapes

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

## MODEL 1 — Linear Regression

### STEP 8 — Create Linear Regression Model

```python
lr_model = LinearRegression()
```

### STEP 9 — Train Linear Regression

```python
lr_model.fit(X_train, y_train)
```

### STEP 10 — Generate Predictions

```python
y_pred_lr = lr_model.predict(X_test)
```

### STEP 11 — Check Predictions

```python
print(y_pred_lr[:5])
```

### STEP 12 — Create Random Forest Model

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
```

### STEP 13 — Train Random Forest

```python
rf_model.fit(X_train, y_train)
```

### STEP 14 — Generate Predictions

```python
y_pred_rf = rf_model.predict(X_test)
```

### STEP 15 — Check Predictions

```python
print(y_pred_rf[:5])
```

## MODEL 3 — Gradient Boosting

### STEP 16 — Create Gradient Boosting Model

```python
gb_model = GradientBoostingRegressor(
    n_estimators=100,
    random_state=42
)
```

### STEP 17 — Train Gradient Boosting

```python
gb_model.fit(X_train, y_train)
```

### STEP 18 — Generate Predictions

```python
y_pred_gb = gb_model.predict(X_test)
```

### STEP 19 — Check Predictions

```python
print(y_pred_gb[:5])
```

### STEP 20 — Save Predictions

## DEEP LEARNING LSTM MODEL IMPLEMENTATION
#### we build and train the LSTM model.

### STEP 1 — Import Libraries

```python
import numpy as np

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

### STEP 2 — Load Processed Arrays

#### Load the arrays you saved during preprocessing.

```python
X_train = np.load("stock_datasets/X_train.npy")

X_test = np.load("stock_datasets/X_test.npy")

y_train = np.load("stock_datasets/y_train.npy")

y_test = np.load("stock_datasets/y_test.npy")
```

### STEP 3 — Check Shapes

```python
print(X_train.shape)

print(X_test.shape)

print(y_train.shape)

print(y_test.shape)
```

### STEP 4 — Build LSTM Model

#### Now create the neural network.

```python
model = Sequential()
```

### STEP 5 — Add First LSTM Layer

```python
model.add(
    LSTM(
        units=50,
        return_sequences=True,
        input_shape=(X_train.shape[1], 1)
    )
)

model.add(Dropout(0.2))
```

### STEP 6 — Add Second LSTM Layer

```python
model.add(
    LSTM(
        units=50,
        return_sequences=False
    )
)

model.add(Dropout(0.2))
```

### STEP 7 — Add Dense Layers

```python
model.add(Dense(units=25))

model.add(Dense(units=1))
```

### STEP 8 — Compile Model

```python
model.compile(
    optimizer="adam",
    loss="mean_squared_error"
)
```

### STEP 9 — View Model Summary

```python
model.summary()
```

### STEP 10 — Train LSTM Model
#### This is the main training step.

```python
history = model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=20,
    validation_data=(X_test, y_test)
)
```

### STEP 11 — Generate Predictions

```python
y_pred_lstm = model.predict(X_test)
```

### STEP 12 — Check Predictions

```python
print(y_pred_lstm[:5])
```

### STEP 13 — Save LSTM Model

```python
model.save("stock_datasets/lstm_model.h5")
```

### STEP 14 — Save Predictions

```python
import pandas as pd

lstm_predictions = pd.DataFrame({
    "Actual": y_test.flatten(),
    "Predicted": y_pred_lstm.flatten()
})

lstm_predictions.to_csv(
    "stock_datasets/lstm_predictions.csv",
    index=False
)
```
