```python
## Import libraries
import yfinance as yf
import pandas as pd
import numpy as np
import time
import os
import requests
```

```python
# CONFIG
STOCKS = ["RELIANCE.NS", "TCS.NS", "INFY.NS", "HDFCBANK.NS", "ICICIBANK.NS"]
START = "2015-01-01"
END = "2026-04-17"

os.makedirs("stock_datasets", exist_ok=True)

## Download data
for stock in STOCKS:
    print(f"\nDownloading {stock}...")

    try:
        df = yf.download(
            stock,
            start=START,
            end=END,
            progress=False,
            auto_adjust=True
        )

        if df.empty:
            print(f" No data: {stock}")
            continue

        # Fix MultiIndex
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.get_level_values(0)

        df = df.reset_index()
        df = df[["Date", "Open", "High", "Low", "Close", "Volume"]]

        df = df.dropna().drop_duplicates()

        file_name = f"stock_datasets/{stock.replace('.NS','')}.csv"
        df.to_csv(file_name, index=False)

        print(f" Saved {file_name} ({len(df)} rows)")

    except Exception as e:
        print(f" Error for {stock}: {e}")

    time.sleep(2)

print("\n DATA COLLECTION COMPLETE")
```
